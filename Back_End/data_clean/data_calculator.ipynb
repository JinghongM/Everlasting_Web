### Redesign checkUnit funciton.
In this part, I redesigned the checkUnit function, which allow us to check unit whatever it's related to weight or height or size.
when we try to process data, just modify the map list below. But we don't need to change the function.

```python
heightMap = {'\"' : 'inches', 'inches' : 'inches', '\'': 'inches', 'cm' : 'cm', 'CM' : 'cm' }
weightMap = {'kg' : 'kg', 'KG' : 'kg', 'lbs' : 'lbs', 'LBS' : 'lbs'}
```

```python
heightRules = {'cm': 1, 'inches': 2.54}
```

```python
def checkUnit(Map, data):
    unit = None
    for key in Map:
        if key in data:
            return Map.get(key)
    return None
```

```python
# Test for checkUnit
checkUnit(heightMap, '1245-2345cm')
```

### Redesign the numberCalculator funtcion
Put the rules in different maps. Modify the map before using this function. And we don't need to change the function.   BTW, this function will only fetch the last number in an expression. For example, "23.5 - 24.5", this function will only use 24.5

```python
def numberCalculator(data, unit, rules):
    return float(re.findall(r'\d+\.?\d+', data)[-1]) * rules[unit]
```

```python
# Test:
numberCalculator('2.5-3.5inches', 'inches', heightRules)
```

### Redesign the calculator function
Function input: two columns, a map contains mapping of units, and a map of rules

```python
def calculator(column1, column2, Map, rules):
    unit = None
    for i in range(len(column1)):
        if type(column1[i]) is str or math.isnan(column1[i]) is False:
            data = str(column1[i])
            tmpUnit = checkUnit(Map, data)
            if tmpUnit is not None:
                unit = tmpUnit
            number = numberCalculator(data, unit, rules)
            if number != 0:
                column2[i] = number
```
